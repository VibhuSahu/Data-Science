# Why you should not use normal List?

- Slow
- Takes memory
- Not efficient

```python
# Using of Array and Marker
# Array are faster Then List
"""
For More Marker characters 
link: https://matplotlib.org/stable/api/markers_api.html
"""

# x - axis
x = np.array([2023, 2024, 2025, 2026])
# y - axis
y = np.array([15, 25, 30, 20])


plot.plot(x,y, 
            marker="h",      # for Dot icon
            ms=10,           # for Dot Size also called as 'markersize'       
            markerfacecolor="#eb4034",  # for Dot color
            markeredgecolor="#757373",  # for edge color
            linestyle="dotted",         # for pattern of the edge for invisible make it none
            linewidth=4)        # Line connecting the points will get thick

plot.show()
```

```python
# Multiple Y axis

# x - axis
x = np.array([2023, 2024, 2025, 2026])
# y - axis
y1 = np.array([8, 18, 28, 38])
y2 = np.array([12, 22, 18, 28])
y3 = np.array([5, 10, 20, 25])

# y1 data getting plot
plot.plot(x,y1, 
            marker="h",      # for Dot icon
            ms=10,           # for Dot Size also called as 'markersize'       
            markerfacecolor="#eb4034",  # for Dot color
            markeredgecolor="#a103fc",  # for edge color
            linestyle="dotted",         # for pattern of the edge for invisible make it none
            linewidth=4)             # Line connecting the points will get thick


# y2 data getting plot
plot.plot(x,y2, 
            marker="h",      # for Dot icon
            ms=10,           # for Dot Size also called as 'markersize'       
            markerfacecolor="#eb4034",  # for Dot color
            markeredgecolor="#0ffc03",  # for edge color
            linestyle="dotted",         # for pattern of the edge for invisible make it none
            linewidth=4)             # Line connecting the points will get thick


# y3 data getting plot
plot.plot(x,y3, 
            marker="h",      # for Dot icon
            ms=10,           # for Dot Size also called as 'markersize'       
            markerfacecolor="#eb4034",  # for Dot color
            markeredgecolor="#fc4903",  # for edge color
            linestyle="dotted",         # for pattern of the edge for invisible make it none
            linewidth=4)             # Line connecting the points will get thick


plot.show()
```

# Bar Chart
    
- Compare categories of data by representing each category with a bar.

```python

# x-axis (classes)
classes = np.array(["Class A", "Class B", "Class C", "Class D"])

# y-axis (certificate holders)
certificates = np.array([25, 40, 30, 50])

# Bar data plotting
plot.bar(classes, certificates)

# style dictionary for lables
lable_style = dict(family="DejaVu Sans")

# Title for the Graph
plot.title("Classes Performance",fontweight="bold", color="#03f4fc", fontsize=29,**lable_style)

#  X and Y Lable
# Lable for x - axis       
plot.xlabel("Mark", fontweight="normal", fontsize=15,**lable_style)

# Lable for x - axis       
plot.ylabel("Year", fontweight="normal", fontsize=15, **lable_style)


# Tick Params
plot.tick_params(axis="both",
                 color="#8cfc03")

# legend
plot.legend()


# y - axis grid lines
plot.grid(axis="y",
          linewidth=0.5,
          color="#768761")

# x - axis grid lines
plot.grid(axis="x",
          linewidth=1,
          color="#768761")

plot.show() 
```

```python
# Multiple Bar in Y - axis

# x-axis (classes)
classes = np.array(["Class A", "Class B", "Class C", "Class D"])

# y-axis (certificate holders)
certificates1 = np.array([5, 10, 20, 25])        # 1st Year
certificates2 = np.array([12, 22, 18, 28])       # 2nd Year
certificates3 = np.array([25, 40, 30, 50])       # 3rd Year

# positions for bars
x = np.arange(len(classes))   # [0,1,2,3]
width = 0.25

# plotting grouped bars
plot.bar(x - width, certificates1, width, color="#e30b0b", label="1st Year")
plot.bar(x,         certificates2, width, color="#97e30b", label="2nd Year")
plot.bar(x + width, certificates3, width, color="#0be3d8", label="3rd Year")

# set class labels
plot.xticks(x, classes)


# style dictionary for lables
lable_style = dict(family="DejaVu Sans")

# Title for the Graph
plot.title("Classes Performance",fontweight="bold", color="#03f4fc", fontsize=29,**lable_style)

#  X and Y Lable
# Lable for x - axis       
plot.xlabel("Classes", fontweight="normal", fontsize=15,**lable_style)

# Lable for x - axis       
plot.ylabel("No. of Certificates", fontweight="normal", fontsize=15, **lable_style)


# Tick Params
plot.tick_params(axis="both",
                 color="#8cfc03")

# legend
plot.legend()


# y - axis grid lines
plot.grid(axis="y",
          linewidth=0.5,
          color="#768761")


plot.show() 
```

# Pie Charts

- Circular chart divided into slices to show percentages of the total.
- Good for visualizing distribution among categories.

```python
# categories
categories = ["Electronics", "Clothing", "Groceries", "Furniture", "Sports"]

# sales data
sales = np.array([35, 25, 20, 10, 10])

# colors
colors = ["#ff6b6b", "#4ecdc4", "#ffe66d", "#1a535c", "#ff9f1c"]


plot.pie(sales, labels=categories,          # Lable names
                autopct="%1.2f%%")          # Lableing percentage on top of chart

plot.show()
```

```python
# Explode

# categories
categories = ["Electronics", "Clothing", "Groceries", "Furniture", "Sports"]

# sales data
sales = np.array([35, 25, 20, 10, 10])

# colors
colors = ["#ff6b6b", "#4ecdc4", "#ffe66d", "#1a535c", "#ff9f1c"]


plot.pie(sales, labels=categories, 
                autopct="%1.2f%%",
                explode=[0,0,0,0.2,0],      # Showing different from the rest of the chart
                shadow=True,                # Giving shadow to chart
                startangle=100)             # Rotation Angle


plot.title("Char Showing Sales Percentage")
plot.tight_layout()
plot.show()

```

# Scatter Plot

- Shows the relationship between two variables
- Helps to identify a correlation (+, -, None)
- Example: Study hours vs Test scores

```python
# X - axis
x = np.array([10.5, 20.2, 30.8, 40.1])

# Y - axis
y = np.array([5.3, 15.7, 25.4, 35.9])

# Plotting a scatter Graph
plot.scatter(x,y)

# Title
plot.title("Test Scores", color="#e30b3e",
                          fontsize=25,
                          fontfamily="monospace",
                          fontweight="bold")

# lable for x - axis                          
plot.xlabel("Hours", color="#949293",
                          fontsize=12,
                          fontfamily="monospace",
                          fontweight="bold")

# lable for y - axis                          
plot.ylabel("Grades Scored", color="#949293",
                          fontsize=12,
                          fontfamily="monospace",
                          fontweight="bold")


plot.show()
```

```python
# Multi Data Scatter plot

# X - axis
x = np.array([10.5, 20.2, 30.8, 40.1])

# Y - axis
y1 = np.array([5.3, 15.7, 25.4, 35.9])      # 1st y - axis
y2 = np.array([6.5, 14.2, 22.8, 30.1])      # 2nd y - axis
y3 = np.array([4.8, 12.9, 20.5, 28.7])      # 3rd y - axis

# Plotting a scatter Graph
# Ploting First data
plot.scatter(x,y1, color="#7af26d",
                  alpha=0.5,
                  s=200,
                  label="Vinay")
# Ploting Second data
plot.scatter(x,y2, color="#f2de6d",
                  alpha=0.5,
                  s=200,
                  label="Kunal")
# Ploting Third data
plot.scatter(x,y3, color="#f2916d",
                  alpha=0.5,
                  s=200,
                  label="Durga")

# Title
plot.title("Test Scores", color="#e30b3e",
                          fontsize=25,
                          fontfamily="monospace",
                          fontweight="bold")

# lable for x - axis                          
plot.xlabel("Hours", color="#949293",
                          fontsize=12,
                          fontfamily="monospace",
                          fontweight="bold")

# lable for y - axis                          
plot.ylabel("Grades Scored", color="#949293",
                          fontsize=12,
                          fontfamily="monospace",
                          fontweight="bold")


plot.legend()
plot.show()
```

# Histograms

- A visual representation of the distribution of quantitative data.
- They group values into bins (intervales)
- and counts how many fall in each ranges>

```python
# Data For Ploting
scores = np.random.normal(loc=80,       # Center of curve
                          scale=10,     # divident
                          size=100)     # Range till 100

# Range Between 0 to 120
scores = np.clip(scores, 0, 120)

# Ploting
plot.hist(scores, bins=10,
                  color="lightgreen",
                  edgecolor="black")


# Title
plot.title("Exam Scores", color="#e30b3e",
                          fontsize=25,
                          fontfamily="monospace",
                          fontweight="bold")

# lable for x - axis                          
plot.xlabel("Score", color="#949293",
                          fontsize=12,
                          fontfamily="monospace",
                          fontweight="normal")

# lable for y - axis                          
plot.ylabel("# of students", color="#949293",
                          fontsize=12,
                          fontfamily="monospace",
                          fontweight="normal")

plot.show()
```

# Subplots

- Figure = The entire canvas
- Ax = A single plot (subplot)

```python
import numpy as np
import matplotlib.pyplot as plt

# Create figure with 2x2 grid
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# [0, 0] Scatter Plot
x1 = np.array([1, 2, 3, 4])
y1 = np.array([10.5, 20.2, 30.8, 40.1])

axes[0, 0].scatter(x1, y1,
                   color="#7af26d",
                   alpha=0.7,
                   s=200)

axes[0, 0].set_title("Scatter Plot")

# [0, 1] Line Plot 
x2 = np.array([1, 2, 3, 4])
y2 = np.array([5, 15, 25, 35])

axes[0, 1].plot(x2, y2,
                marker="o",
                color="#03a9f4",
                linewidth=2)

axes[0, 1].set_title("Line Plot")

# [1, 0] Bar Chart
classes = ["A", "B", "C", "D"]
values = [20, 35, 30, 25]

axes[1, 0].bar(classes, values,
               color=["#ff6b6b", "#4ecdc4", "#ffe66d", "#1a535c"])

axes[1, 0].set_title("Bar Chart")

# [1, 1] Pie Chart
sizes = [40, 30, 20, 10]
labels = ["Electronics", "Clothing", "Groceries", "Others"]

axes[1, 1].pie(sizes,
               labels=labels,
               autopct="%1.1f%%",
               colors=["#ff9f1c", "#2ec4b6", "#e71d36", "#011627"])

axes[1, 1].set_title("Pie Chart")



# Adjust layout
plt.tight_layout()

plt.show()
```
